# II Reading and arranging data section:

### w(N=1)

```python
# Nierika BAO + JLA

how_many_chains1=len(chain_root_baojla_nierika)

R1=[]


for i in chain_root_baojla_nierika:
    
    #these lines read .h_info (for R-1 values) and .v_info (for names of parameters) 
    r1=np.genfromtxt(i+'.h_info', skip_header=1, delimiter=':',dtype=None)
    nam1=np.genfromtxt(i+'.v_info', skip_header=1, delimiter=':',dtype=None)
    
    #these lines convert the string 'r1'...
    v1=r1[0,1].tostring()                       #array with just R-1numbers.(numbers as strings)
    f1=v1.split('\t')                            #float list with R-1 numbers as floats
    f1.pop()                                    #removes last element of list (because it appears "empty")
    
    #these line converts 'nam' into a list of names
    names1=nam1[:,0]
    
    #print f
    #print names
    
    R1.append(f1)

#print(R1)

for i in range(len(R1)):
    print 'R-1 values for N=1 with BAO + JLA chain', i, 'are:'
    print R1[i]
```

```python
# Nierika CMB+BAO+SN

how_many_chains2=len(chain_root_planck_nierika)
#for i in chain_root:
#    print i

R2=[]


for j in chain_root_planck_nierika:
    
    #these lines read .h_info (for R-1 values) and .v_info (for names of parameters) 
    r2=np.genfromtxt(j+'.h_info', skip_header=1, delimiter=':',dtype=None)
    nam2=np.genfromtxt(j+'.v_info', skip_header=1, delimiter=':',dtype=None)
    
    #these lines convert the string 'r1'...
    v2=r2[0,1].tostring()                       #array with just R-1numbers.(numbers as strings)
    f2=v2.split('\t')                           #float list with R-1 numbers as floats
    f2.pop()                                    #removes last element of list (because it appears "empty")
    
    #these line converts 'nam' into a list of names
    names2=nam2[:,0]    
    R2.append(f2)


for j in range(len(R2)):
    print 'R-1 values for N=1 chain with Planck + BAO + JLA', j, 'are:'
    print R2[j]
```
